# XGBoost
## 1.1 通用数据准备

```python
import pandas as pd
import numpy as np
import pickle
import xgboost as xgb
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import log_loss, roc_auc_score, auc, roc_curve
from sklearn.model_selection import train_test_split
```

```python
dataset = pd.read_csv("Basic_feature.csv", index_col=0)
dataset
```

## 1.2 有目的地填补缺失值

```python
temp1 = ['u3','u4','u5','u10','u11','u12','u13','um11','um12','c11','c12','m4','m5','m6','m7','m8','m9',]
    
for i in temp1:
    dataset[i]=dataset[i].fillna(-1)

dataset = dataset.replace({np.nan:0})
dataset
```

## 1.3 替换分类变量，防止意外出错；初步分类X和Y

```python
dataset.drop_duplicates(inplace=True)
dataset.replace({"Label":{-1:2}},inplace = True) #-1类换为2
dataset_y = dataset[['Label']]
dataset_x = dataset.drop(['User_id','Merchant_id', 'Coupon_id','Label'],axis=1)
```

## 1.4 PCA后分训练集和测试集

```python
from sklearn import datasets
from sklearn.decomposition import PCA


features = ['c1','c2','c3','c4','c6','c8','c9','c10','c11','c12','m1','m2','m3','m4','m5','m6','m7','m8','m9','u1','u2','u3','u4','u5','u6','u7','u8','u9','u10','u11','u12','u13','um0','um1','um2','um3','um4','um5','um6','um7','um8','um9','um10','um11','um12']
X = dataset_x[features]

pca = PCA(n_components=20) # we select PCA algorithm with 2 dimensions
pca = pca.fit(X) # same as 1.3
z = pca.transform(X)

PCA_Features = pd.DataFrame(z)
PCA_Features.columns=(['pca_z1','pca_z2','pca_z3','pca_z4','pca_z5','pca_z6','pca_z7','pca_z8','pca_z9','pca_z10','pca_z11','pca_z12','pca_z13','pca_z14','pca_z15','pca_z16','pca_z17','pca_z18','pca_z19','pca_z20'])
PCA_Features
```

```python
print(PCA_Features.shape,dataset_y.shape)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(
    PCA_Features,
    #dataset_x, 
    dataset_y, test_size=0.3, random_state=42)
```

## 1.5 不平衡数据处理

```python
from imblearn.over_sampling import ADASYN, SMOTE, RandomOverSampler

sampler = RandomOverSampler(shrinkage=1, random_state=0)
X_train_ROS, Y_train_ROS = sampler.fit_resample(X_train, Y_train)
```

```python
#选取数据集
#SMOTE和ADASYN太慢了，用ROS代替训练先
X_train = X_train_ROS
Y_train = Y_train_ROS
```

```python
Y_train.value_counts('Label')
```

```python
def myauc(test):
    testgroup = test.groupby(['coupon_id'])
    aucs = []
    for i in testgroup:
        tmpdf = i[1] 
        if len(tmpdf['label'].unique()) != 2:
            continue
        fpr, tpr, thresholds = roc_curve(tmpdf['label'], tmpdf['pred'], pos_label=1)
        aucs.append(auc(fpr,tpr))
    return np.average(aucs)
```

## 2. XGB

```python
dataTrain = xgb.DMatrix(X_train,label=Y_train)
dataTrain_Test = xgb.DMatrix(X_test)

params={'booster':'gbtree',
        'objective': 'multi:softmax',#输出多元结果
        'eval_metric':'merror',
        'gamma':10,
        'min_child_weight':10,
        'max_depth':2,
        'lambda':100,
        'subsample':1,
        'colsample_bytree':0.7,
        'colsample_bylevel':0.7,
        'eta': 0.01,
        'tree_method':'gpu_hist',
        'seed':0,
        'nthread':12,
        #'early_stopping_rounds':
        'early_stopping_rounds': 1,
        'num_class': 3 #objective 为 muti:softmax是需要定义的分类数
        }

watchlist = [(dataTrain,'train')]
#eval_set = [(X_train, Y_train), (X_test, Y_test)]
model = xgb.train(params,dataTrain,num_boost_round=100, #次数少的时候就已经过拟合了
                  #eval_metric="error", eval_set=eval_set,
                 )

model.save_model('/xgbmodel')
model=xgb.Booster(params)
model.load_model('/xgbmodel') 
```

```python
#predict test set 
model=xgb.Booster()
model.load_model('/xgbmodel') 
predict = Y_test
predict['pict'] = model.predict(dataTrain_Test)
predict
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import plot_confusion_matrix
from sklearn.metrics import confusion_matrix
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

y_true = predict['Label']
y_pred = predict['pict']

cm = confusion_matrix(y_true, y_pred)

disp = ConfusionMatrixDisplay(confusion_matrix = cm)
disp.plot(cmap = 'Blues_r')

print(classification_report(y_true, y_pred))
```
