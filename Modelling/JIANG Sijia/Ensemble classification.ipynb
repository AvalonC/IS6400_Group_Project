```python
from sklearn.model_selection import train_test_split # this function provides a single "Hold-Out" Validation.
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score #similar to MAE, we use accuracy_score evaluation metric.
import pandas as pd
from pandas import DataFrame
import numpy as np
```

```python
data = pd.read_csv('Basic_feature.csv')
data.head(5)
```

```python
data.drop(columns='Unnamed: 0', inplace=True)
```

#### PCA

```python
import matplotlib.pyplot as plt

from sklearn import datasets
from sklearn.decomposition import PCA 
```

```python
data = data.fillna(value=-1)
data
```

```python
data.drop_duplicates(inplace=True)
data
```

```python
data["Label"]
```

```python
data.columns=['User_id', 'Merchant_id', 'Coupon_id', 'c1', 'c2', 'c3', 'c4', 'c5',
       'c6', 'c7', 'c8', 'c9', 'c10', 'c11', 'c12', 'Label', 'm1', 'm2', 'm3',
       'm4', 'm5', 'm6', 'm7', 'm8', 'm9', 'u1', 'u2', 'u3', 'u4', 'u5', 'u6',
       'u7', 'u8', 'u9', 'u10', 'u11', 'u12', 'u13', 'um0', 'um1', 'um2',
       'um3', 'um4', 'um5', 'um6', 'um7', 'um8', 'um9', 'um10', 'um11',
       'um12']

features=['c1', 'c2', 'c3', 'c4', 'c5',
       'c6', 'c7', 'c8', 'c9', 'c10', 'c11', 'c12', 'm1', 'm2', 'm3',
       'm4', 'm5', 'm6', 'm7', 'm8', 'm9', 'u1', 'u2', 'u3', 'u4', 'u5', 'u6',
       'u7', 'u8', 'u9', 'u10', 'u11', 'u12', 'u13', 'um0', 'um1', 'um2',
       'um3', 'um4', 'um5', 'um6', 'um7', 'um8', 'um9', 'um10', 'um11',
       'um12']
X=data[features]
```

```python
X
```

```python
pca = PCA(n_components=20) #20 dimensions
pca = pca.fit(X) 
z = pca.transform(X)
```

```python
F = pd.DataFrame(columns=['z1', 'z2', 'z3', 'z4','z5', 'z6', 'z7', 'z8','z9', 'z10', 'z11', 'z12','z13', 'z14', 'z15', 'z16','z17', 'z18', 'z19', 'z20'])
F['z1']=z[:,0] 
F['z2']=z[:,1]
F['z3']=z[:,2]
F['z4']=z[:,3]
F['z5']=z[:,4]
F['z6']=z[:,5]
F['z7']=z[:,6]
F['z8']=z[:,7]
F['z9']=z[:,8]
F['z10']=z[:,9]
F['z11']=z[:,10]
F['z12']=z[:,11]
F['z13']=z[:,12]
F['z14']=z[:,13]
F['z15']=z[:,14]
F['z16']=z[:,15]
F['z17']=z[:,16]
F['z18']=z[:,17]
F['z19']=z[:,18]
F['z20']=z[:,19]
F.head()
```

```python
data['Label']
```

```python
Y = data['Label']
X = F
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.4, random_state=1) 
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import BaggingClassifier
np.random.seed(1)
clf1 = BaggingClassifier(DecisionTreeClassifier(max_depth=2),n_estimators=50) # number estimators is the key parameter
clf1 = clf1.fit(X_train, Y_train)
Y_predTrain = clf1.predict(X_train)
Y_predTest = clf1.predict(X_test)

print('Training accuracy is %f' % accuracy_score(Y_train, Y_predTrain))
      
print('Validation accuracy is %f' %accuracy_score(Y_test, Y_predTest))
```
