# 4. User interaction with merchant feature

In practical applications, it is important to explore the usage habits of consumer coupon at each merchant, so we conduct user interaction with merchant feature.

```python
# Number of
temp=offline_notNull.groupby(['User_id','Merchant_id']).size().reset_index(name='um0')
X=temp
X
```

```python
# The number of times the user receives the merchant's coupons 
temp = offline_notNull.groupby(['User_id','Merchant_id']).size().reset_index(name='um1')
X = pd.merge(X, temp, how='left', on=['User_id','Merchant_id'])
X
```

```python
#The number of times the user does not use after receiving the merchant's coupons
temp = coupon_no_consume.groupby(['User_id', 'Merchant_id']).size().reset_index(name='um2')
X = pd.merge(X, temp, how='left', on=['User_id', 'Merchant_id'])
X
```

```python
# The number of times the user uses after receiving the merchantâ€™s coupons 
temp = coupon_consume.groupby(['User_id','Merchant_id']).size().reset_index(name='um3')
temp
```

```python
X = pd.merge(X, temp, how='left', on=['User_id','Merchant_id'])
X
```

```python
# Usage rate after users receive coupons from merchants (times of using/times of not using)
X['um4'] = X.um3 / X.um1
X
```

```python
# The ratio of users' non-use times for each merchant to the total number of users' non-use times 
temp = coupon_no_consume.groupby('User_id').size().reset_index(name='temp')
X = pd.merge(X, temp, how='left', on='User_id')
X['um5'] = X.um2 / X.temp
X.drop(columns='temp', inplace=True)
X
```

```python
# Total number of times the user has consumed in the store 
temp = offline[offline.Date.notnull()].groupby(['User_id', 'Merchant_id']).size().reset_index(name='um6')
X = pd.merge(X, temp, how='left', on=['User_id', 'Merchant_id'])
X
```

```python
date_received_isnull_date_notnull = offline[offline.Date_received.isnull() & offline.Date.notnull()]
date_received_isnull_date_notnull
```

```python
# Ordinary consumption times of users in the store 
temp = date_received_isnull_date_notnull
temp = temp.groupby(['User_id', 'Merchant_id']).size().reset_index(name='um7')
X = pd.merge(X, temp, how='left', on=['User_id', 'Merchant_id'])
X
```

```python
date_received_notnull = offline[offline.Date_received.notnull()]
date_received_notnull
```

```python
# Times of users receiving coupons at each merchant on the same day 
temp = date_received_notnull.groupby(['User_id','Merchant_id']).size().reset_index(name='um8')
X = pd.merge(X, temp, how='left', on=['User_id','Merchant_id'])
X
```

```python
# The number of merchants that a user has received coupons from 
temp = offline_notNull.groupby(['User_id', 'Merchant_id']).size().reset_index()
temp = temp.groupby('User_id').size().reset_index(name='um9')
X = pd.merge(X, temp, how='left', on='User_id')
X
```

```python
# The number of merchants that a user has verified coupons UM10
temp = coupon_consume.groupby(['User_id', 'Merchant_id']).size()
temp = temp.groupby('User_id').size().reset_index(name='um10')
X = pd.merge(X, temp, how='left', on='User_id')
X
```

```python
#  Ratio of the number of different merchants whose coupons have been written off by users to all different merchants 
X['um11'] = X.um10 / X.um9
X
```

```python
User_feature = pd.read_csv('User_feature.csv', header=0)
print(User_feature.shape[0])
User_feature
```

```python
# The average number of coupons written off by users for each merchant 
X['um12'] = User_feature.u1 / X.um9
X['um12']
```
